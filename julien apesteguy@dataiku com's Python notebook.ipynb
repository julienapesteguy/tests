```python
import dataiku
from dataiku import pandasutils as pdu
import pandas as pd
```

```python
# Example: load a DSS dataset as a Pandas dataframe
mydataset = dataiku.Dataset("SUBSCRIPTION")
df = mydataset.get_dataframe()
```

```python
df
```

```python
from datetime import datetime
```

```python
def getDate(x):
    if(x != x):
        return
    return datetime.fromtimestamp(x)
```

```python
newdf = df[["ID", "CUSTOMER_ID", "STATUS", "SUBSCRIPTION_ITEMS", "BILLING_PERIOD_UNIT", "NEXT_BILLING_AT", "CREATED_AT", "UPDATED_AT", "STARTED_AT", "CANCELLED_AT", "TRIAL_END"]].iloc[:15]
```

```python
date = {"NEXT_BILLING_AT","CREATED_AT","UPDATED_AT","STARTED_AT","CANCELLED_AT","TRIAL_END"}

for col in date:
    newdf[col] = newdf[col].apply(getDate)
```
